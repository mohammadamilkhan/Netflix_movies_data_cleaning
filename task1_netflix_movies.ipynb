```python
# 1. Check for missing values and handle them
missing_values = df.isnull().sum()
df = df.fillna({
    'director': 'Unknown',
    'cast': 'Unknown',
    'country': 'Unknown',
    'date_added': '01-01-1900'  # Placeholder for missing dates
})
```

```python

# 2. Check for duplicates and remove them
df = df.drop_duplicates()
```

```python
# 3. Standardize text values
df['gender'] = df.get('gender', pd.Series(["Unknown"] * len(df))).str.lower().str.strip()
df['country'] = df['country'].str.title().str.strip()
```

```python
# 4. Convert date formats to consistent dd-mm-yyyy
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.strftime('%d-%m-%Y')
```

```python
# 5. Rename column headers to be clean and uniform
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# 6. Check and fix data types
# Manually correcting some expected types
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
df['date_added'] = pd.to_datetime(df['date_added'], format='%d-%m-%Y', errors='coerce')
```
